### Understand about features

```python
list(df.columns)
```

- These columns typically represent attributes in a dataset for mobile apps, such as those from the Google Play Store or similar app repositories. Here's a description of each column:

    - 1. App: The name of the application (e.g., "WhatsApp", "Facebook").
    
    - 2. Category:The category the app belongs to, such as "Social", "Games", "Productivity", or "Education".
    
    - 3. Rating:The average user rating for the app, usually on a scale of 1 to 5.
    
    - 4. Reviews:The total number of user reviews submitted for the app.
    
    - 5. Size:The file size of the app, often measured in MB (megabytes). Some datasets may include "Varies with device" if the size is not fixed.
    
    - 6. Installs:The total number of installations or downloads of the app, typically represented in ranges (e.g., "1,000+", "10,000+").
    
    - 7. Type:Indicates whether the app is free or paid. Typical values are "Free" or "Paid".
    
    - 8. Price:The cost of the app. For free apps, this is typically 0, while for paid apps, this column shows the price.
    
    - 9. Content Rating:The age group for which the app is appropriate (e.g., "Everyone", "Teen", "Mature 17+").
    
    - 10. Genres:The genres associated with the app, providing more specific categorization than "Category" (e.g., "Action", "Puzzle", "Casual").
    
    - 11. Last Updated:The most recent date when the app was updated.
    
    - 12. Current Ver (Current Version):The version number of the app as of the most recent update.

    - 13. Android Ver (Android Version):The minimum Android operating system version required to run the app (e.g., "4.0 and up", "5.1 and up").

### 1. Bar Charts

- 1.	Number of Apps by Category:
   - o	Objective: To visualize the distribution of the number of apps across different categories.
   -o	Conclusion: The bar chart shows the count of apps for each category, revealing which categories have the most apps.


```python
#  Number of Apps by Category


# Group by 'Category' and count the number of apps in each category
category_counts = df['Category'].value_counts()

# Create a bar chart
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Number of Apps by Category')
plt.xticks(rotation=90)
plt.show()
```

- 2.	Free vs Paid Apps:
    - o	Objective: To analyze the distribution of free and paid apps.
    - o	Conclusion: The bar chart shows a significant proportion of apps are free, with a smaller percentage being paid apps


```python
# b. Free vs Paid Apps
price_counts = df['Price'].apply(lambda x: 'Free' if x == '0' else 'Paid').value_counts()

sns.barplot(x=price_counts.index, y=price_counts.values)
plt.title('Free vs Paid Apps')
plt.show()
```

- 3.	Free vs Paid Apps by Category (Stacked Bar Chart):
    - o	Objective: To explore the number of free and paid apps within each category.
    - o	Conclusion: Stacked bar charts show that some categories, like "Lifestyle" or "Games", have a high proportion of free apps, while others have more paid apps


```python
### . Stacked Bar Charts
# c. Free vs Paid Apps by Category
category_free_paid = pd.crosstab(df['Category'], df['Price'].apply(lambda x: 'Free' if x == '0' else 'Paid'))

category_free_paid.plot(kind='bar', stacked=True, figsize=(12,8))
plt.title('Free vs Paid Apps by Category')
plt.ylabel('Number of Apps')
plt.show()
```

- 4.	Average rating by category
    - Objective: To compare the average ratings across different categories and highlight the performance distribution.

    - Conclusion: The diverging bar chart reveals that some categories have significantly higher ratings than others, with notable disparities between positive and negative feedback across categories.

```python
### . Diverging Bar Chart
#  A diverging bar chart can help you compare positive and negative values on the same scale
category_avg_rating = df.groupby('Category')['Rating'].mean().sort_values()

category_avg_rating.plot(kind='barh', color='b', figsize=(10, 6))
plt.title('Average Rating by Category (Diverging Bar)')
plt.show()
```

- 5.	Free vs Paid Apps by Category (Alternative Stacked Bar Chart):
    - o	Objective: To compare the free and paid apps across categories using another stacked bar chart.
    - o	Conclusion: Similar to the previous plot, this shows how different categories have varying proportions of free vs paid apps


```python
### Stacked Bar Chart - Free vs Paid Apps by Category
# Count the number of free vs paid apps per category
free_paid = df.groupby(['Category', 'Type']).size().unstack()

# Plot a stacked bar chart
free_paid.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Free vs Paid Apps by Category')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=90)
plt.show()
```
